```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Spotify_songs.csv')
```

```python
data=data.dropna()
```

```python
data=data.drop_duplicates()
```

```python
data
```

```python
data.shape
```

```python
data['specifity']=data['playlist_name']+data['playlist_genre']+data['playlist_subgenre']
```

```python
data=data.drop('playlist_name', axis=1).reset_index(drop=True)
```

```python
data=data.drop('playlist_genre', axis=1).reset_index(drop=True)
data=data.drop('playlist_subgenre', axis=1).reset_index(drop=True)
```

```python
data
```

```python
plt.scatter(data['playlist_genre'],data['playlist_subgenre'])
```

```python
colonnes_specifiques = ['track_name', 'track_artist','specifity']
new_data = data[colonnes_specifiques]

```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

def stopword(text):
 text = ' '.join([word for word in str(text).split() if word not in stopwords.words("english")])
 return text
new_data['specifity']=new_data['specifity'].apply(stopword)
```

```python
import nltk
nltk.download('punkt')
import re
from nltk.tokenize import word_tokenize
#convertir maj en min
new_data['specifity']=new_data['specifity'].str.lower()
#supprimer les ponctuation
def pop_ponctuation(text):
    var=re.compile('[^\w\s]')
    return var.sub(r'',text)
new_data['specifity']=new_data['specifity'].apply(pop_ponctuation)
#supprimer les caractères uniques 
def pop_car_unique(text):
    words=word_tokenize(text)
    text= ' '.join([word for word in words if len(word)>1])
    return text
new_data['specifity']=new_data['specifity'].apply(pop_car_unique)
```

```python
new_data
```

```python
nltk.download('wordnet')
#tokenization
def tokenize(text):
    text=word_tokenize(str(text))
    return text
new_data['specifity']=new_data['specifity'].apply(tokenize)
```

```python
new_data
```

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
def lemmatize(text):
    lemmatize=WordNetLemmatizer()
    text = ' '.join([lemmatize.lemmatize(i) for i in str(text).split()])
    return text
new_data['specifity']=new_data['specifity'].apply(lemmatize)
```

```python
import pandas as pd
new_data=pd.to_csv('final_dataset.csv')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tv = TfidfVectorizer(analyzer='word',stop_words='english')
vec= tv.fit_transform(new_data['specifity'])
similarity = cosine_similarity(vec)
```

```python
def recommend(song):
    if song in new_data['track_name'].unique():
        song_index = new_data[new_data['track_name'] == song].index[0]
        distances = similarity[song_index]
        song_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:5]
        for i in song_list:
            print(new_data.iloc[i[0]].track_name)
    else:
        print(f"Aucune chanson correspondant '{song}' n'a été trouvée dans le dataset.")
```

```python
recommend('Payphone')
```
